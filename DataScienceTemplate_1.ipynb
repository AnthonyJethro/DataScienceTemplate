# General Preprocessing 

```python
# check unique value from each column

data['emp_length'].unique()
```

```python
# replace/remove STRING
data['emp_length_int'] = data['emp_length'].str.replace('\+ years', '')
data['emp_length_int'] = data['emp_length_int'].str.replace('< 1 year', str(0))
data['emp_length_int'] = data['emp_length_int'].str.replace('n/a',  str(0))
data['emp_length_int'] = data['emp_length_int'].str.replace(' years', '')
data['emp_length_int'] = data['emp_length_int'].str.replace(' year', '')
type(data['emp_length_int'][0])
```

```python
# change value to INT

data['emp_length_int'] = pd.to_numeric(data['emp_length_int'])
type(data['emp_length_int'][0])
```

```python
#convert date and time


data['earliest_cr_line_date'] = pd.to_datetime(data['earliest_cr_line'], format = '%b-%y')
type(data['earliest_cr_line_date'][0])
```

```python
#create new column that represent date/time
pd.to_datetime('2017-12-01') - data['earliest_cr_line_date']
#time is set to certain time
```

```python
#round up data

data['mths_since_earliest_cr_line'] = round(pd.to_numeric((pd.to_datetime('2017-12-01') - data['earliest_cr_line_date']) / np.timedelta64(1, 'M')))
data['mths_since_earliest_cr_line'].describe()
```

```python
# replace abnormal data with something reasonable
#this case minus data will be assume have the same value (very old customer) and assume have max value

data['mths_since_earliest_cr_line'][data['mths_since_earliest_cr_line'] < 0] = data['mths_since_earliest_cr_line'].max()
```

# Preprocess discreet variable

```python
#data some data can be preprocesed into dummy without statistical method
pd.get_dummies(data['grade'])
```

```python
#use this to made it easier
pd.get_dummies(data['grade'], prefix = 'grade', prefix_sep = ':')
```

```python
# do the same thing with other data
data_dummies = [pd.get_dummies(data['grade'], prefix = 'grade', prefix_sep = ':'),
                     pd.get_dummies(data['sub_grade'], prefix = 'sub_grade', prefix_sep = ':'),
                     pd.get_dummies(data['home_ownership'], prefix = 'home_ownership', prefix_sep = ':'),
                     pd.get_dummies(data['verification_status'], prefix = 'verification_status', prefix_sep = ':'),
                     pd.get_dummies(data['loan_status'], prefix = 'loan_status', prefix_sep = ':'),
                     pd.get_dummies(data['purpose'], prefix = 'purpose', prefix_sep = ':'),
                     pd.get_dummies(data['addr_state'], prefix = 'addr_state', prefix_sep = ':'),
                     pd.get_dummies(data['initial_list_status'], prefix = 'initial_list_status', prefix_sep = ':')]

data_dummies
```

```python
#combine all dummies data frame to form 1 dummies data frame
data_dummies = pd.concat(data_dummies, axis = 1)
data_dummies.head()
```

```python
#combine original dataframe with dummies data frame

data = pd.concat([data, data_dummies], axis = 1)
```

# Check for missing values

```python
#check if there is a null value in data frame
data.isnull()
```

```python
#check total number of null data per columns
pd.options.display.max_rows = None
data.isnull().sum()
```

```python
#fill null/na 
#there are other method like useing Average/max/min value from all column 

data['total_rev_hi_lim'].fillna(data['funded_amnt'], inplace = True)
```

```python
#fill na with average value with mean

data['annual_inc'].fillna(data['annual_inc'].mean(), inplace=True)
```

```python
#fill na with 0

data['mths_since_earliest_cr_line'].fillna(0, inplace=True)
data['acc_now_delinq'].fillna(0, inplace=True)
data['total_acc'].fillna(0, inplace=True)
data['pub_rec'].fillna(0, inplace=True)
data['open_acc'].fillna(0, inplace=True)
data['inq_last_6mths'].fillna(0, inplace=True)
data['delinq_2yrs'].fillna(0, inplace=True)
data['emp_length_int'].fillna(0, inplace=True)
```

# Data preparation to find label (assumption)

```python
#In this case user that have bad repuataion is someone that havent pay loan in 90 days
data['loan_status'].unique()
```

```python
data['loan_status'].value_counts()
```

```python
data['loan_status'].value_counts() / data['loan_status'].count()
```

```python
# made assumption label from case
#data that charged off, default, does not meet criteria, late more than 31 days will be assume have bad reputation
# 0 mean default and 1 have a good reputation

data['good_bad'] = np.where(data['loan_status'].isin(['Charged Off', 'Default',
                                                               'Does not meet the credit policy. Status:Charged Off',
                                                               'Late (31-120 days)']), 0, 1)
```
